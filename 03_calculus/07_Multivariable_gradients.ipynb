## 1. From Derivative to Gradient

Single variable:
$$
f(x) = x^2 \quad \Rightarrow \quad \frac{df}{dx} = 2x
$$

Multiple variables:
$$
f(x, y) = x^2 + y^2
$$

Partial derivatives:
$$
\frac{\partial f}{\partial x} = 2x, \quad
\frac{\partial f}{\partial y} = 2y
$$

Gradient (vector of partial derivatives):
$$
\nabla f(x, y) =
\begin{bmatrix}
\frac{\partial f}{\partial x} \\
\frac{\partial f}{\partial y}
\end{bmatrix}
$$


```python
import numpy as np

def f(x,y):
  return x**2 + y**2

def grad_f(x,y):
  return np.array([2*x, 2*y])

point = (1.0, -2.0)
grad_f(*point)
```

## 2. Geometric Meaning

- Gradient points in direction of **steepest ascent**
- Magnitude = rate of increase
- Gradient is **perpendicular to level curves**


```python
# visualizing gradients

import matplotlib.pyplot as plt

x = np.linspace(-3, 3, 20)
y = np.linspace(-3, 3, 20)
X, Y = np.meshgrid(x, y)

Z = f(X, Y)

U = 2*X
V = 2*Y

plt.contour(X, Y, Z)
plt.quiver(X, Y, U, V)
plt.xlabel("x")
plt.ylabel("y")
plt.title("Gradient field: f(x, y)")
plt.show()
```

## 3. Gradient Descent in Multiple Dimensions

Update rule:
$$
\mathbf{w}_{t+1} = \mathbf{w}_t - \eta \nabla L(\mathbf{w}_t)
$$

- Weights are vectors
- Gradient is a vector
- Subtraction happens component-wise


```python
# 2-D Gradient Descent

w = np.array([2.0, -3.0])
lr = 0.1

trajectory = []

for _ in range(20):
  w = w -lr * grad_f(w[0], w[1])
  trajectory.append(w.copy())

trajectory[-1]
```

```python
# Optimization Path

traj = np.array(trajectory)

plt.contour(X, Y, Z)
plt.plot(traj[:, 0], traj[:, 1], 'ro-')
plt.xlabel("x")
plt.ylabel("y")
plt.title("Gradient Descent: f(x, y)")
plt.show()
```

## 4. Key Takeaways

- Gradient generalizes derivative to multiple variables
- Optimization happens in vector space
- Core concept behind all ML training
