# Calculus — Gradient Descent

**Author:** Murigu Gitonga  
**Goal:** Understand gradient descent as an optimization algorithm that uses
derivatives to minimize loss functions in machine learning.

---

> Gradient descent is how machines **learn**.
> It uses derivatives to iteratively reduce error.






## 1. Intuition

- ML models are trained by minimizing a **loss function**
- Loss measures how wrong the model is
- Gradient tells us the **direction of steepest increase**
- To minimize loss, move in the **opposite direction**

$$
\text{New parameter} = \text{Old parameter} - \eta \cdot \text{Gradient}
$$

where $\eta$ (eta) is the learning rate


## 2. Example Loss Function

We start with a simple quadratic loss:

$$
L(w) = (w - 3)^2
$$

- The minimum occurs at $w = 3$
- Derivative:

$$
\frac{dL}{dw} = 2(w - 3)
$$


```python
import numpy as np

# Loss function
def loss(w):
  return (w-3) ** 2

# derivative of loss
def grad(w):
  return 2 * (w -3)

# gradient descent
w = 0.0
lr = 0.1
steps = 20

history = []

for i in range(steps):
   w = w - lr * grad(w)
   history.append([w, loss(w)])

history

```

```python
# visualizing optimization
import matplotlib.pyplot as plt
w_vals = np.linspace(-1, 6, 100)
loss_vals = loss(w_vals)

plt.plot(w_vals, loss_vals)
plt.scatter([x[0] for x in history], [x[1] for x in history])
plt.xlabel('w')
plt.ylabel('loss')
plt.title('Gradient Descent Optimization Path')
plt.show()
```

## 3. Connection to Machine Learning

- Parameters (weights) are updated using gradient descent
- For many parameters → gradients become vectors
- For neural networks → gradients computed using backpropagation
- Variants:
  - Batch Gradient Descent
  - Stochastic Gradient Descent (SGD)
  - Mini-batch Gradient Descent


```python
# dataset -> linear regression
x = np.array([1, 2, 3, 4])
y = np.array([2, 4, 6, 8]) # y = 2x

#mode: y = wx
w = 0.0
lr = 0.01

def loss(w):
  y_pred = w * x
  return np.mean((y - y_pred) ** 2)

def grad(w):
  y_pred = w * x
  return -2 * np.mean(x * (y - y_pred))

# Training loop
for epoch in range(50):
  w = w -lr * grad(w)

w
```

```python
# loss curve cell
losses = []

w = 0.0

for epoch in range(50):
  w = w - lr * grad(w)
  losses.append(loss(w))

plt.plot(losses)
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.title('Loss Curve: training loss over time')
plt.show()
```

## 4. Key Takeaways

- Gradient descent minimizes loss using derivatives
- Learning rate controls step size
- Too large → divergence
- Too small → slow learning
- Core of all ML optimization
