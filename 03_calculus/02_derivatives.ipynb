# Calculus - Derivatives

**Objective:** Understand derivatives as rates of change and connect them directly to loss minimization in machine learning.



## 1. Intuition

A **derivative** measures how fast a function changes when its input changes.

Think of it as:
- The **slope** of a curve
- The **steepness** at a point
- Sensitivity of output to input changes

If the derivative is:
- Positive → function increases
- Negative → function decreases
- Zero → flat (minimum or maximum)


## 2. Mathematical Definition

The derivative of a function $f(x)$ at point $x$ is defined as:

$$
f'(x) = \lim_{h \to 0} \frac{f(x + h) - f(x)}{h}
$$

This formula captures:
- Change in output  
- Divided by change in input  
- As the change becomes very small


```python
import numpy as np

def numerical_derivative(f, x, h = 1e-5):
  return(f(x + h) - f(x)) / h

def f(x):
  return x**2

x = 3.0
numerical_derivative(f, x)

```

## 3. Analytical vs Numerical Derivatives

- **Analytical derivative**: exact formula (e.g., $f'(x) = 2x$)
- **Numerical derivative**: approximation using finite differences

In ML:
- Backprop computes analytical gradients
- Numerical derivatives are used for **gradient checking**


```python
import matplotlib.pyplot as plt

x = np.linspace(-5, 5, 100)

y = x**2
dy_dx = 2 * x

plt.figure()
plt.plot(x, y, label = "f'(x) - x^2")
plt.xlabel("x")
plt.legend()
plt.title("A function and its derivate")
plt.show()
```

## 4. Derivatives in Machine Learning

In ML, we define a **loss function**:

$$
L(w) = (y - \hat{y})^2
$$

We compute:

$$
\frac{dL}{dw}
$$

This tells us:
- How changing the weight affects the error
- Which direction reduces the loss


```python
def loss(w, x=1.0, y=2.0):
  y_pred = w * x
  return (y - y_pred) ** 2

w = 0.5
numerical_derivative(loss, w)
```

## 5. Interpretation

- Large derivative → steep slope → big change needed
- Small derivative → near minimum → small updates
- Zero derivative → optimal point (minimum)

Training stops when derivatives are close to zero.


## 6. Key Takeaways

- Derivatives measure rate of change
- They indicate direction and magnitude of updates
- Optimization relies entirely on derivatives
- Numerical derivatives help validate gradients
- No derivatives → no learning
