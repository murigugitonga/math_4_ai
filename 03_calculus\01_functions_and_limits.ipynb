## 1. Intuition

Defining functions

```python
import numpy as np

def linear_function(x, w=2, b=1):
  return w* x + b

x = np.array([-2, -1, 0, 1, 2])
y = linear_function(x)

x, y
```

Visualizing functions

```python
import matplotlib.pyplot as plt

x = np.linspace(-10, 10, 100)
y = linear_function(x)

plt.figure()
plt.plot(x,y)
plt.xlabel("x")
plt.ylabel("f(x)")
plt.title("Linear function f(x) = 2x + 1")
plt.show()
```

## 3. Limits - Intuition

A **limit** describes what a functions output approaches as the input approaches a certain value.

\[
\lim{ x \to a} f(x)
\]

key takeaway:
- The function does not need to be defined at \(x = a \)
- Of importance is what happens *near* \(a\)

Limits help define
- Continuity
- Smoothness
- Change (derivatives come next)

```python
def f(x):
  return (x**2 - 1) /(x-1)

x = np.linspace(0.5, 1.5, 400)
y = f(x)

plt.figure()
plt.plot(x,y)
plt.axvline(1, linestyle = '--')
plt.xlabel("x")
plt.ylabel("f(x)")
plt.title("limit as x -> 1")
plt.show()
```

## 4. Why we need limits in AI

Limits form the foundations of:

- Derivates
- Gradients
- Optimization

In training:

- Critical question: *What happens to the loss if parameters change slightly?*
- Limits sufficiently answer that quesion.
